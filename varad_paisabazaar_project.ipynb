# **Project Summary -**

For this project, the goal was to analyze customer data from Paisabazaar to uncover patterns and potential red flags that might indicate banking fraud. Some of the important features examined were Age, SSN, Occupation, Annual Income, Credit Utilization Ratio, Payment Behavior, and Credit Score. ## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Ensure df is a fresh copy of the original dataset to prevent issues from prior modifications.
df = dataset.copy()

# Converting age column from float to integer
df['Age'] = df['Age'].astype(int)

#checking for duplicate values
duplicates = df.duplicated().sum()
print("Number of duplicate row:", duplicates)

#checking for null values
null_values = df.isnull().sum().sum()
print("Number of null values:", null_values)

# verifying after cleaning
print("Shape after droping the duplicates and the null values:",df.shape)
````

after checking the duplicates and null values. we will remove the outliers

````python
# removing the outliers in the 'Annual_Income' column
Q1 = df['Annual_Income'].quantile(0.25)
Q3 = df['Annual_Income'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Annual_Income'] >= (Q1 - 1.5 * IQR)) & (df['Annual_Income'] <= (Q3 + 1.5 * IQR))]
````

````python
# Example: Creating a new feature 'Income_per_Account'
df['Income_per_Account'] = df['Annual_Income'] / df['Num_Bank_Accounts']

# Example: Creating a feature 'Credit_Score_Level'
df['Credit_Score_Level'] = df['Credit_Score'].apply(lambda x: 'High' if isinstance(x, int) and x > 700 else 'Low')
````

After that we calculated removed the outliers in 'Annual_Income' using IQR method.we then Converted 'credit_score' to numeric and also removed the nan values in 'credit_score' for better accuracy.
